```r
library(ggplot2)
library(reshape2)
library(data.table)
```

```r
# read csv file
tidy_cor <- fread("./correlation_matrix_tidylong.csv")
```

```r
# Create the heatmap
ggplot(tidy_cor, aes(x = Feature1, y = Feature2, fill = Correlation)) +
  geom_tile() +
  scale_fill_gradient2(low = "blue", high = "red", mid = "white", 
                       midpoint = 0, limit = c(-1, 1), space = "Lab") +
  theme_minimal() +
  theme(axis.text.x = element_blank(), axis.text.y = element_blank()) +
  labs(title = "Correlation Matrix Heatmap", fill = "Correlation")
```
